```python
print(json.dumps(["python","C++","C"]))
print(json.dumps(("Java","html","hv")))
print(json.dumps({"Name":"Arun","Age":25}))
```

```python
import json
student = '{"Name": "Prakash","Age":25,"Rollno": 56,"CGPA": 8.6,"Phonenumber": "9976770500","Languages":["Python","Java"]}'
dic=json.loads(student)
print(json.dumps(dic,indent=6,sort_keys=True))
```

```python
dictionary = {
    "Name": "Vijay",
    "Rollno": 56,
    "CGPA": 8.6,
    "Phonenumber": "9976770500",
    "Langauges":'Python'
} 
json_object = json.dumps(dictionary, indent=4)
with open("vijay.json", "w") as outfile:
    outfile.write(json_object)
```
