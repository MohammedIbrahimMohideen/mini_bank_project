Logistic Regression
It is one of the most popular ML algorithms, which comes under the Supervised Learning technique . It is used for predicting the categorical dependent variable using a given set of independent variables.

Logistic regression predicts the output of a categorical dependent variable. Therefore the outcome must be a categorical or discrete value.It can be either Yes or No, o or 1, true or false

Logistic regression is much similar to the Linear Regression except that how they are used. Linear Regression is used for solving Regression problems, whereas Logistic regression is used for solving the classification problems.

Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
c=pd.read_csv('D:/Ibbibanz/Python/DigitalAd_dataset (1).csv')
c
```

Summarize Dataset

```python
print('Shape of the Data set:',c.shape)
print('*'*20)
print(c.head())
```

```python
X=c.iloc[:,:-1]
X
```

```python
X=c.iloc[:,:-1].values
X
```

```python
Y=c.iloc[:,-1]
Y
```

```python
Y=c.iloc[:,-1].values
Y
```

Splitting Dataset into Train & Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=1)
```

```python
x_train
```

```python
x_test
```

Feature Scaling\ We scale our data to make all the features contribute equally to the result\ Fit_Transform - fit method: is calculating the mean and variance of each of the features present in our data - Training Purpose\ Transform - Trasform method is transforming all the features using the respective mean and variance - Testing Purpose

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

Predicting, Wheather new customer with Age & Salary will Buy or Not

```python
age=int(input('Enter New Customer Age:'))
salary=int(input('Enter New Customer Salary:'))
newcus=[[age,salary]]
result=model.predict(sc.transform(newcus))
print(result)
```

```python
age=int(input('Enter New Customer Age:'))
salary=int(input('Enter New Customer Salary:'))
newcus=[[age,salary]]
result=model.predict(sc.transform(newcus))
print(result)
if result==1:
    print('Customer Will Buy The Product')
else:
    print('Customer Wont Buy The Product')
```

Prediction for all Test Data

```python
y_test
```

```python
y_pred=model.predict(x_test)
y_pred
```

Evaluating Model -CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_pred)
print('Confusion Matrix:\n',cm)
```

Accuracy = TP + TN / TP + TN + FP + FN

```python
(56 + 24) / (56 + 24 + 16 + 4)
```

```python
acc_score=accuracy_score(y_test,y_pred)
print(f'Accuracy of the Model:,{acc_score*100}%')
```
